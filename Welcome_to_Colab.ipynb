```
class Student:
  def __init__(self, name, age, grade):
    self.name=name
    self.age=age
    self.grade=grade

s1=Student('Usman Khan',22,'A')
print(s1.name,s1.age,s1.grade)
```

```
class BankAccount:
  def __init__(self, holder, balance):
    self.holder=holder
    self.balance=balance

  def deposit(self, amount):
    self.balance=self.balance+amount
    return self.balance
  def withdraw(self,credit):
    self.balance=self.balance-credit
    return self.balance

acc=BankAccount('Usman',20000)
acc.deposit(3000)
acc.withdraw(7000)
print(acc.holder, acc.balance)
```
