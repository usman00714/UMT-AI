```
class Student:

  def __init__(self, name, marks):
    self.name=name
    self.marks=marks


  def has_passed(self):
    print("Passed")
    return self.marks>=60
  def has_failed(self):
    self.marks<60
    return print("FAIL!!!")


s1=Student('Ali', 76)
s2=Student('Usman', 49)

print(s1.name,s1.has_passed())
print(s2.name,s2.has_passed())
```
