```
class Student:
  def __init__(self, name, marks):
    self.name=name
    self.marks=marks

  def has_passed(self):
    print("Passed")
    return self.marks>=60
  def has_failed(self):
    self.marks<60
    return print("FAIL!!!")


s1=Student('Ali', 76)
s2=Student('Usman', 49)

print(s1.name,s1.has_passed())
print(s2.name,s2.has_passed())
```

```
class Car:
  def __init__(self, brand, model):
    self.brand=brand
    self.model=model

  def full_name(self):
    return f"{self.brand} {self.model}"

class Electric_car(Car):
  def __init__(self,brand, model, battery):
    super().__init__(brand,model)
    self.battery=battery

byd=Electric_car('BYD','SEAL',45000)
print(byd.brand, byd.full_name(),byd.model, byd.battery)





c1=Car('Honda', 'Civic')
c2=Car('BMW','I5')
print(c1.brand,c2.brand)
print(c1.full_name())
print(c2.model, c1.model)


```
